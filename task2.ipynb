```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "/content/Au_nanoparticle_dataset.csv"
data = pd.read_csv(file_path)
```

```python
selected_columns = ["N_total", "N_bulk", "N_surface", "R_avg"]
new_df = data[selected_columns]
```

```python
print(new_df.head(20))
```

```python
feature_stats = new_df.describe()
```

```python
new_df.hist(layout=(1, 4), figsize=(15, 3))
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(new_df)
plt.show()
```

```python
g = sns.PairGrid(new_df)
g.map_upper(plt.hist2d)  # bivariate histogram
g.map_diag(sns.histplot, kde=True)  # histogram with kde
g.map_lower(sns.kdeplot)  # bivariate kde
plt.show()
```
